```python
import geedim
import geemap
import ee
ee.Authenticate()
ee.Initialize()
```

```python
roi = ee.FeatureCollection('USDOS/LSIB_SIMPLE/2017').filter(ee.Filter.eq('country_na', 'Thailand'))
crs = 'EPSG:4326'
```

```python
nicfi = (ee.ImageCollection('projects/planet-nicfi/assets/basemaps/asia')
         .filter(ee.Filter.date('2021-03-01', '2021-07-01'))
         .filterBounds(roi)
         .median()
         .select(['R', 'G', 'B'])
         .clip(roi))
```

```python
Map = geemap.Map(center=[13.7, 100.5], zoom=6)  # Adjust center and zoom for Thailand
# Add the linear fit result for Thailand to the map
vis = {"bands":["R","G","B"],"min":64,"max":5454,"gamma":2};
Map.addLayer(nicfi, vis, 'Planet Lab 2021 (Thailand)')
# Display the map
Map
```

```python
geemap.download_ee_image(
            nicfi,
            scale=5,
            filename='/content/planet_2021.tif',
            dtype='float32',
            region=roi.geometry(),
            crs=crs
        )
```
