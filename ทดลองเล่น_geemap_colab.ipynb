## ตัวเลือกแผนที่ฐาน

เลือกแผนที่ฐานจากรายการแบบเลื่อนลงและเพิ่มลงในแผนที่

```python
m = geemap.Map()
m.add("basemap_selector")
m
```

## Layer Manager

สลับช่องทำเครื่องหมายเพื่อแสดงหรือซ่อนเลเยอร์ ลากและเลื่อนแถบเลื่อนเพื่อเปลี่ยนระดับความโปร่งใสของเลเยอร์

```python
m = geemap.Map(center=(13, 101), zoom=6)
dem = ee.Image('USGS/SRTMGL1_003')
thailand = ee.FeatureCollection('USDOS/LSIB_SIMPLE/2017').filter(ee.Filter.eq('country_na', 'Thailand'))
vis_params = {
    'min': 0,
    'max': 4000,
    'palette': ['006633', 'E5FFCC', '662A00', 'D8D8D8', 'F5F5F5'],
}
m.add_layer(dem, vis_params, 'SRTM DEM')
m.add_layer(thailand, {}, "Thailand")
m.add('layer_manager')
m
```

## Inspector

คลิกบนแผนที่เพื่อสืบค้นข้อมูล Earth Engine ในตำแหน่งที่ต้องการ

```python
m = geemap.Map(center=(13, 101), zoom=6)
dem = ee.Image('USGS/SRTMGL1_003')
landsat7 = ee.Image('LANDSAT/LE7_TOA_5YEAR/1999_2003')
thailand = ee.FeatureCollection('USDOS/LSIB_SIMPLE/2017').filter(ee.Filter.eq('country_na', 'Thailand'))
vis_params = {
    'min': 0,
    'max': 4000,
    'palette': ['006633', 'E5FFCC', '662A00', 'D8D8D8', 'F5F5F5'],
}
m.add_layer(dem, vis_params, 'SRTM DEM')
m.add_layer(
    landsat7.clip(thailand),
    {'bands': ['B4', 'B3', 'B2'], 'min': 20, 'max': 350, 'gamma': 2.0},
    'Landsat 7',
)
#m.add_layer(thailand, {}, "ประเทศไทย")
m.add('inspector')
m
```

## Layer Editor

คุณสามารถเปลี่ยนพารามิเตอร์การแสดงภาพของข้อมูล Earth Engine ได้โดยใช้โปรแกรมแก้ไขเลเยอร์

### Single-band image

```python
m = geemap.Map(center=(13, 101), zoom=6)
dem = ee.Image('USGS/SRTMGL1_003')
vis_params = {
    'min': 0,
    'max': 4000,
    'palette': ['006633', 'E5FFCC', '662A00', 'D8D8D8', 'F5F5F5'],
}
m.add_layer(dem, vis_params, 'SRTM DEM')
m.add('layer_editor', layer_dict=m.ee_layers['SRTM DEM'])
m
```

### Multi-band image
ทดลองผสมแบนด์ดูได้ จะสังเกตได้ว่าข้อมูลภาพที่นำมาเป็นตัวอย่างคือข้่อมูล Landsat-7 โดยธรรมชาติของข้อมูลจะมี stripes เป็นเส้นๆ แต่ในตัวอย่างได้ทำการ destriping เรียบร้อยแล้ว

```python
m = geemap.Map(center=(13, 101), zoom=6)
landsat7 = ee.Image('LANDSAT/LE7_TOA_5YEAR/1999_2003')
thailand = ee.FeatureCollection('USDOS/LSIB_SIMPLE/2017').filter(ee.Filter.eq('country_na', 'Thailand'))
m.add_layer(
    landsat7.clip(thailand),
    {'bands': ['B4', 'B3', 'B2'], 'min': 20, 'max': 200, 'gamma': 2.0},
    'Landsat 7',
)
m.add('layer_editor', layer_dict=m.ee_layers['Landsat 7'])
m
```

### Feature collection

```python
m = geemap.Map(center=(13, 101), zoom=6)
thailand = ee.FeatureCollection("TIGER/2018/States")
m.add_layer(thailand, {}, "Thailand")
m.add('layer_editor', layer_dict=m.ee_layers['Thailand'])
m
```

## Draw control

คุณสามารถวาดรูปร่างบนแผนที่ได้โดยใช้ตัวควบคุมการวาด คุณลักษณะที่วาดไว้จะถูกแปลงเป็นวัตถุ Earth Engine โดยอัตโนมัติ ซึ่งสามารถเข้าถึงได้ด้วยวิธีใดวิธีหนึ่งต่อไปนี้:

หากต้องการคืนคุณลักษณะที่วาดล่าสุดเป็น ee.Geometry() ให้ใช้: m._draw_control.last_geometry
หากต้องการส่งคืนฟีเจอร์ที่วาดทั้งหมดเป็น ee.Feature() ให้ใช้: m._draw_control.last_feature
หากต้องการส่งคืนคุณสมบัติที่วาดทั้งหมดเป็น ee.FeatureCollection() ให้ใช้: m._draw_control.collection

```python
m = geemap.Map(center=(13, 101), zoom=6)
dem = ee.Image('USGS/SRTMGL1_003')
vis_params = {
    'min': 0,
    'max': 4000,
    'palette': 'terrain',
}
m.add_layer(dem, vis_params, 'SRTM DEM')
m.add('layer_manager')
m
```
